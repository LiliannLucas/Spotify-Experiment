```
from fancyimpute import KNN, NuclearNormMinimization, SoftImpute, BiScaler
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import HTML
```

```
cm = sns.light_palette("green", as_cmap=True)
```

```
def highlight_col(s):
  color = 'white' if s.is_integer() else 'darkorange'
  return 'color: %s' % color
```

```
data = pd.read_csv("/content/sample.csv", index_col='User')
```

```
X = data.to_numpy()
X = X.astype(np.float16)
```

```
# randome mask data as unknown, now some user's song-preference is unknown. To demostrate the formular, made some simple sample
missing_mask = missing_mask = np.random.rand(*X.shape) < 0.1
X_incomplete = X.copy()
X_incomplete = X_incomplete
X_incomplete[missing_mask] = np.nan
data_incomplete = pd.DataFrame(data=X_incomplete, dtype=np.float16)
data_incomplete.columns = ['Song1', ' Song2',' Song3', ' Song4',' Song5']
```

```
X_filled_knn = KNN(k=3).fit_transform(X_incomplete)
X_filled_nnm = NuclearNormMinimization().fit_transform(X_incomplete)
X_incomplete_normalized = BiScaler().fit_transform(X_incomplete)
X_filled_softimpute = SoftImpute().fit_transform(X_incomplete_normalized)
```

```
# check errors
nnm_mse = ((X_filled_nnm[missing_mask] - X[missing_mask]) ** 2).mean()
print("nnm MSE: %f" % nnm_mse)
softImpute_mse = ((X_filled_softimpute[missing_mask] - X[missing_mask]) ** 2).mean()
print("SoftImpute MSE: %f" % softImpute_mse)
knn_mse = ((X_filled_knn[missing_mask] - X[missing_mask]) ** 2).mean()
print("knnImpute MSE: %f" % knn_mse)
```

```
df_X_filled_nnm = pd.DataFrame(data=X_filled_nnm)
df_X_filled_nnm.columns = ['Song1', ' Song2',' Song3', ' Song4',' Song5']
```
